```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize global style
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline

# List of required datasets
DATASETS = {
    'games': 'games.csv',
    'plays': 'plays.csv',
    'players': 'players.csv',
    'week_data': 'week_data.csv'
}

def load_datasets():
    """Load all required datasets with proper error handling"""
    data = {}
    try:
        for name, path in DATASETS.items():
            data[name] = pd.read_csv(path)
            print(f"‚úÖ Successfully loaded {name} dataset ({len(data[name])} rows)")
        return data
    except FileNotFoundError as e:
        print(f"‚ùå Critical Error: Missing file {e.filename}")
        raise
    except Exception as e:
        print(f"‚ö†Ô∏è Unexpected error: {str(e)}")
        raise

# Load all datasets
data = load_datasets()
games = data['games']
plays = data['plays']
players = data['players']
week_data = data['week_data']
```

```python
# Create inspection summary table
def inspect_dataset(df, name):
    """Generate dataset inspection summary"""
    return pd.DataFrame({
        'Dataset': [name],
        'Rows': [df.shape[0]],
        'Columns': [df.shape[1]],
        'Missing Values': [df.isna().sum().sum()],
        'Duplicate Rows': [df.duplicated().sum()],
        'Memory Usage (MB)': [round(df.memory_usage(deep=True).sum()/(1024*1024), 2)]
    })

# Generate inspection report for all datasets
inspection_report = pd.concat([
    inspect_dataset(games, 'games'),
    inspect_dataset(plays, 'plays'),
    inspect_dataset(players, 'players'),
    inspect_dataset(week_data, 'week_data')
], ignore_index=True)

print("üìã Dataset Inspection Report:")
display(inspection_report)
```

```python
def analyze_columns(df, dataset_name):
    """Detailed column analysis for each dataset"""
    print(f"\nüîç {dataset_name.upper()} COLUMN ANALYSIS:")

    # Column types summary
    dtype_counts = df.dtypes.value_counts()
    print(f"\nüìä Data Types:\n{dtype_counts}")

    # Missing values analysis
    missing_values = df.isna().sum()
    missing_percent = (missing_values/len(df)*100).round(2)
    missing_df = pd.DataFrame({
        'Missing Values': missing_values,
        'Percentage (%)': missing_percent
    }).sort_values('Missing Values', ascending=False)

    print(f"\n‚ùì Missing Values Summary (Top 5):")
    display(missing_df.head(5))

    # Memory optimization check
    mem_before = df.memory_usage(deep=True).sum()
    optimized_df = df.copy()
    for col in optimized_df.select_dtypes(include='object'):
        optimized_df[col] = optimized_df[col].astype('category')
    mem_after = optimized_df.memory_usage(deep=True).sum()

    print(f"\nüíæ Memory Optimization Potential: {round((mem_before - mem_after)/1024**2, 2)} MB")

# Analyze each dataset
analyze_columns(games, 'games')
analyze_columns(plays, 'plays')
analyze_columns(players, 'players')
analyze_columns(week_data, 'week_data')
```

# **Data Cleaning & Preprocessing**

```python
# Fixing plays dataset
plays['preSnapVisitorScore'] = plays['preSnapVisitorScore'].fillna(plays['preSnapVisitorScore'].median())
plays['typeDropback'] = plays['typeDropback'].fillna(plays['typeDropback'].mode()[0])
plays['gameClock'] = plays['gameClock'].fillna('00:00')

# Fixing week_data dataset
week_data['o'] = week_data['o'].fillna(week_data['o'].median())
week_data['position'] = week_data['position'].fillna('UNKNOWN')
week_data['jerseyNumber'] = week_data['jerseyNumber'].fillna(-1)

# Verifying missing values after fixing
print("Plays Missing Values After Fix:\n", plays.isna().sum().head())
print("Week_Data Missing Values After Fix:\n", week_data.isna().sum().head())
```

```python
# Optimize games dataset
games = games.astype({
    'homeTeamAbbr': 'category',
    'visitorTeamAbbr': 'category',
    'gameTimeEastern': 'category'
})

# Optimize plays dataset
plays = plays.astype({
    'personnelO': 'category',
    'personnelD': 'category',
    'typeDropback': 'category'
})

# Optimize players dataset
players = players.astype({
    'collegeName': 'category',
    'position': 'category',
    'displayName': 'category'
})

# Optimize week_data dataset
week_data = week_data.astype({
    'position': 'category',
    'team': 'category'
})
```

```python
# Convert height from feet-inches to inches
def convert_height(height):
    if isinstance(height, str) and '-' in height:
        ft, inches = height.split('-')
        return int(ft) * 12 + int(inches)
    return float(height)

players['height'] = players['height'].apply(convert_height)

# Calculate player age
players['birthDate'] = pd.to_datetime(players['birthDate'], errors='coerce')
players['age'] = (pd.to_datetime('2025-02-23') - players['birthDate']).dt.days // 365

# Extract game year from gameDate
games['gameYear'] = pd.to_datetime(games['gameDate']).dt.year
```

```python
games.to_csv('cleaned_games.csv', index=False)
plays.to_csv('cleaned_plays.csv', index=False)
players.to_csv('cleaned_players.csv', index=False)
week_data.to_csv('cleaned_week_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Debug: Print actual column names
print("Column names in dataset:", players.columns)

# Convert 'birthDate' to datetime and calculate 'age'
if "birthDate" in players.columns:
    players["birthDate"] = pd.to_datetime(players["birthDate"], errors='coerce')
    players["age"] = datetime.now().year - players["birthDate"].dt.year
else:
    print("Error: 'birthDate' column not found!")

# Convert height to numeric (fixing non-numeric values)
players["height"] = pd.to_numeric(players["height"], errors='coerce')

# Fill missing values
players["height"].fillna(players["height"].median(), inplace=True)
players["age"].fillna(players["age"].median(), inplace=True)

# Set up figure
plt.figure(figsize=(12, 5))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(players["age"], bins=20, kde=True, color='blue')
plt.title("Age Distribution")

# Weight Distribution
plt.subplot(1, 2, 2)
sns.histplot(players["weight"], bins=20, kde=True, color='red')
plt.title("Weight Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove outliers using Interquartile Range (IQR)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal
players_cleaned = remove_outliers(players, "age")
players_cleaned = remove_outliers(players_cleaned, "weight")

# Set up figure size
plt.figure(figsize=(10, 5))

# Age Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(x=players_cleaned["age"], color="royalblue")  # Fixed color
plt.title("Age Boxplot")

# Weight Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=players_cleaned["weight"], color="crimson")  # Fixed color
plt.title("Weight Boxplot")

plt.show()
```
